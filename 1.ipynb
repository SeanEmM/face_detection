```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test.jpg')
```

```python
plt.imshow(img)
```

```python
conv_img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(conv_img1)
```

```python
face_cas = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cas
```

```python
faces = face_cas.detectMultiScale(conv_img1,minNeighbors=16)
```

```python
faces
```

```python
for x,y,w,h in faces:
    cv2.rectangle(conv_img1,(x,y),(x+w,y+h),(0,0,255),20)
```

```python
plt.imshow(conv_img1)
```

```python
video = cv2.VideoCapture('video.mp4')
while True:
    success,img = video.read()
    if not success:
        print("End of video")
        break
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_cas.detectMultiScale(gray,minNeighbors=10)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
    cv2.imshow("Image",img)
    if cv2.waitKey(1) & 0XFF==113:
        break
video.release()
cv2.destroyAllWindows()
```
